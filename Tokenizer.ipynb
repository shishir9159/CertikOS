```python
import re

def process_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    processed_lines = []
    for line in lines:
        line = line.strip()
        if not line:  # Skip empty lines
            continue

        if "=" not in line:
            # should have allowed "+" and "=" and then check the line
            if re.search(r'[^\u0980-\u09FF\s\n]', line):
                continue
            else:
              processed_lines.append(line)
        else:
            parts = line.split("=")
            if len(parts) == 2:
                if "+" in parts[1]:
                  parts[0],parts[1]=parts[1],parts[0]
                processed_lines.append("=".join(parts).strip())
            else:
                print("skipping")
              # processed_lines.append(line)
    return "\n".join(processed_lines)

file_path = '/content/sample_data/raw-dataset.txt'
processed_content = process_file(file_path)
len(processed_content)
print(processed_content)
```

```python
def write_string_to_file(filename: str, content: str):
    with open(filename, "w", encoding="utf-8") as file:
        file.writelines(line + "\n" for line in content.split("\n"))

def count_lines_in_file(filename: str) -> int:
    with open(filename, 'r', encoding='utf-8') as file:
        return sum(1 for _ in file)

write_string_to_file("output.txt", processed_content)
filename = "output.txt"
line_count = count_lines_in_file(filename)
print(f"Number of lines: {line_count}")
```
